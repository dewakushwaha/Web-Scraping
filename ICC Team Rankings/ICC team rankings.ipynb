### Importing libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

### Collecting data from URL

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/t20i'
```

```python
r = requests.get(url).text
soup = BeautifulSoup(r,'lxml')
table = soup.find('table', attrs={'class':'table'})
```

```python
teamlist=[]
for tr in table('tr'):
    for td in table('td'):
        teamlist.append(td.text.replace('\n',"").strip())
    break
        
```

```python
teamlist
```

### Reshaping 

```python
array = np.array(teamlist)
```

```python
reshaped = array.reshape((72,5))
```

```python
reshaped
```

### Converting to Dataframe

```python
df = pd.DataFrame(reshaped, columns =['POS','TEAM','MATCHES','POINTS','RATING']) 
```

```python
df.head()
```

```python
df.to_csv("T20I rankings.csv")
```
